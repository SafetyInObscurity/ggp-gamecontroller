Comparing the moves chosen using the true distribution against the moves that would have been chosen without knowledge.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import statistics
np.random.seed(42)
import os

# Display and sorting
# Found code from stackoverflow
from IPython.display import display_html
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
import re
def sorted_alphanumeric(data):
    convert = lambda text: int(text) if text.isdigit() else text.lower()
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ] 
    return sorted(data, key=alphanum_key)

# Read in the csv files
match_id = 'blind_breakthrough_5x5_CHEAT_4x4ophyperb_16CHEAT_1backtrackDepth_squared'
num_games = 50

# Read in op_move_distribution
op_move_dist = list()
for idx in range(0, num_games):
    op_move_dist_instance = list()
    folder_name = '../testOutput/opponentModelling/smaller_games/4x4_ophyperb_squared/op_move_distribution/' + match_id + '_' + str(idx)
    for name in sorted_alphanumeric(os.listdir(folder_name)):
        file_name = folder_name + '/' + name
        op_move_dist_instance.append(pd.read_csv(file_name, names=['move','value']))
    op_move_dist.append(op_move_dist_instance)
    
# Read in orig_move_distribution
orig_move_dist = list()
for idx in range(0, num_games):
    orig_move_dist_instance = list()
    folder_name = '../testOutput/opponentModelling/smaller_games/4x4_ophyperb_squared/orig_move_distribution/' + match_id + '_' + str(idx)
    for name in sorted_alphanumeric(os.listdir(folder_name)):
        file_name = folder_name + '/' + name
        orig_move_dist_instance.append(pd.read_csv(file_name, names=['move','value']))
    orig_move_dist.append(orig_move_dist_instance)

for idx, (op_move, orig_move) in enumerate(zip(op_move_dist[0], orig_move_dist[0])):
    print(str(idx))
    display_side_by_side(op_move, orig_move)
```

It is worth considering how many different predictions were made

```python
op_best_move_list = list()
op_best_value_list = list()
orig_best_move_list = list()
orig_best_value_list = list()

# Iterate through games, movelists and individual moves to compare
for game_idx, (op_move_game, orig_move_game) in enumerate(zip(op_move_dist, orig_move_dist)):
    op_best_move_game = list()
    op_best_value_game = list()
    orig_best_move_game = list()
    orig_best_value_game = list()
#     print('game: ' + str(game_idx))
    for list_idx, (op_move_list, orig_move_list) in enumerate(zip(op_move_game, orig_move_game)):
#         print('move: ' + str(list_idx))
        op_best_value = -100
        op_best_move = None
        orig_best_value = -100
        orig_best_move = None
        for move_idx, (op_move, orig_move) in enumerate(zip(op_move_list.iterrows(), orig_move_list.iterrows())):
#             print('row: ' + str(move_idx))
#             print(op_move[1]['move'] + ' -> ' + str(op_move[1]['value']))
#             print(orig_move[1]['move'] + ' -> ' + str(orig_move[1]['value']))
            if(op_move[1]['value'] > op_best_value):
                op_best_value = op_move[1]['value']
                op_best_move = op_move[1]['move']
            if(orig_move[1]['value'] > orig_best_value):
                orig_best_value = orig_move[1]['value']
                orig_best_move = orig_move[1]['move']
        op_best_move_game.append(op_best_move)
        op_best_value_game.append(op_best_value)
        orig_best_move_game.append(orig_best_move)
        orig_best_value_game.append(orig_best_value)
    op_best_move_list.append(op_best_move_game)
    op_best_value_list.append(op_best_value_game)
    orig_best_move_list.append(orig_best_move_game)
    orig_best_value_list.append(orig_best_value_game)

# Count number of times op_move and orig_move differ
move_count = 0
mismatch_count = 0
for op_best_move_game, orig_best_move_game in zip(op_best_move_list, orig_best_move_list):
    for op_best_move, orig_best_move in zip(op_best_move_game, orig_best_move_game):
        move_count += 1
#         print(op_best_move + ' v ' + orig_best_move)
        if(op_best_move != orig_best_move):
            mismatch_count += 1
#             print('no match')
            
print('Mismatched ' + str(mismatch_count) + ' out of ' + str(move_count) + f' moves for a total of {((mismatch_count/move_count)*100):.2f}%')
```
